```python
import re 
from bs4 import BeautifulSoup

def limpiar_precio(texto_precio):
    return int("".join(re.findall(r'\d+', texto_precio)))

def obtener_categoria(titulo):
    titulo_min = titulo.lower()

    if "mando" in titulo_min or "dualsense" in titulo_min:
        return "Consola + Mando"
    elif "juego" in titulo_min or "bundle" in titulo_min:
        return "Consola + Juego"
    elif "consola" in titulo_min or "ps5" in titulo_min:
        return "Consola Estándar"
    else:
        return "No identificado"
    
    
```
