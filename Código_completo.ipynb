```python

##PROFESORA, ESTE ES EL CÓDIGO COMPLETO, TODOS LOS INTEGRANTES HEMOS PARTICIPADO PERO PARA NO CONFUNDIRNOS DECIDIMOS CADA UNO CREAR UNA RAMA
## Por ende este archivo es solo para que usted pueda visualizar el código completo y los resultados, gracias por su comprensión


import re
from bs4 import BeautifulSoup

# =====================================================================
# FUNCIONES BASE (Paradigma Estructurado)
# =====================================================================
def limpiar_precio(texto_precio):
    """Limpia el texto (ej. 'S/ 2.450') y retorna un número entero (2450)."""
    return int("".join(re.findall(r'\d+', texto_precio)))

def obtener_categoria(titulo):
    """
    Analiza el título y clasifica el producto.
    Si no encuentra palabras clave claras, retorna 'No identificado'.
    """
    titulo_min = titulo.lower()

    if "mando" in titulo_min or "dualsense" in titulo_min:
        return "Consola + Mando"
    elif "juego" in titulo_min or "bundle" in titulo_min:
        return "Consola + Juego"
    elif "consola" in titulo_min or "ps5" in titulo_min:
        return "Consola Estándar"
    else:
        return "No identificado"

# =====================================================================
# FUENTE 1: SIMULACIÓN DE MERCADO LIBRE (Catálogo con ruido)
# =====================================================================
html_mercado_libre = """
<div class="contenedor-productos">
    <div class="producto"><h2 class="titulo-prod">PlayStation 5 Consola de 825GB Standard</h2><span class="precio-prod">S/ 2.450</span></div>
    <div class="producto"><h2 class="titulo-prod">Consola Sony PlayStation 5 Slim 1TB</h2><span class="precio-prod">S/ 2.599</span></div>
    <div class="producto"><h2 class="titulo-prod">Mando Inalámbrico Dualsense Ps5 Blanco</h2><span class="precio-prod">S/ 320</span></div>
    <div class="producto"><h2 class="titulo-prod">PlayStation 5 Digital Edition Slim</h2><span class="precio-prod">S/ 2.200</span></div>
    <div class="producto"><h2 class="titulo-prod">Sony PS5 Standard Edición Física Chasis D</h2><span class="precio-prod">S/ 2.499</span></div>
    <div class="producto"><h2 class="titulo-prod">PlayStation 5 + Juego Spider-Man 2 Bundle</h2><span class="precio-prod">S/ 2.750</span></div>
    <div class="producto"><h2 class="titulo-prod">Consola PS5 1TB Digital Edition Color Blanco</h2><span class="precio-prod">S/ 2.150</span></div>
    <div class="producto"><h2 class="titulo-prod">Playstation 5 Standard 825gb - Versión Japonesa</h2><span class="precio-prod">S/ 2.399</span></div>
    <div class="producto"><h2 class="titulo-prod">Juego Mortal Kombat 1 Ps5 Físico</h2><span class="precio-prod">S/ 250</span></div>
    <div class="producto"><h2 class="titulo-prod">PS5 Consola Edición Disco God Of War Ragnarok</h2><span class="precio-prod">S/ 2.800</span></div>
    <div class="producto"><h2 class="titulo-prod">Consola PlayStation 5 Slim Edición Estándar 1TB</h2><span class="precio-prod">S/ 2.650</span></div>
    <div class="producto"><h2 class="titulo-prod">PS5 Standard 825GB + Mando Extra DualSense</h2><span class="precio-prod">S/ 2.850</span></div>
    <div class="producto"><h2 class="titulo-prod">Auriculares Inalámbricos Pulse 3d Ps5</h2><span class="precio-prod">S/ 450</span></div>
    <div class="producto"><h2 class="titulo-prod">Sony PlayStation 5 Consola Digital 825 GB</h2><span class="precio-prod">S/ 2.299</span></div>
    <div class="producto"><h2 class="titulo-prod">Consola PS5 Standard + FC 24 Bundle Original</h2><span class="precio-prod">S/ 2.790</span></div>
    <div class="producto"><h2 class="titulo-prod">PlayStation 5 Consola 1TB Slim Digital Edition</h2><span class="precio-prod">S/ 2.350</span></div>
    <div class="producto"><h2 class="titulo-prod">PS5 825GB Edición Standard Blanca Nuevo Sellado</h2><span class="precio-prod">S/ 2.480</span></div>
    <div class="producto"><h2 class="titulo-prod">Consola PlayStation 5 (PS5) - Lector de Discos</h2><span class="precio-prod">S/ 2.500</span></div>
    <div class="producto"><h2 class="titulo-prod">PS5 Consola 825GB Edición Digital CFI-1215B</h2><span class="precio-prod">S/ 2.180</span></div>
    <div class="producto"><h2 class="titulo-prod">Funda De Silicona Para Mando Ps5</h2><span class="precio-prod">S/ 35</span></div>
    <div class="producto"><h2 class="titulo-prod">PlayStation 5 Slim 1TB Edición Estándar + 2 Juegos</h2><span class="precio-prod">S/ 2.950</span></div>
    <div class="producto"><h2 class="titulo-prod">Sony Consola PS5 Lector Discos 825gb</h2><span class="precio-prod">S/ 2.449</span></div>
    <div class="producto"><h2 class="titulo-prod">PlayStation 5 Consola Digital 825gb Blanco</h2><span class="precio-prod">S/ 2.249</span></div>
    <div class="producto"><h2 class="titulo-prod">Consola PS5 Slim 1TB Disco D Chasis Nuevo</h2><span class="precio-prod">S/ 2.620</span></div>
    <div class="producto"><h2 class="titulo-prod">PS5 Edición Estándar 825gb + Horizon Forbidden West</h2><span class="precio-prod">S/ 2.700</span></div>
    <div class="producto"><h2 class="titulo-prod">Base De Carga Para Mando Ps5</h2><span class="precio-prod">S/ 120</span></div>
    <div class="producto"><h2 class="titulo-prod">Consola de Videojuegos PlayStation 5 Slim 1TB</h2><span class="precio-prod">S/ 2.550</span></div>
</div>
"""

# =====================================================================
# FUENTE 2: SIMULACIÓN DE FACEBOOK MARKETPLACE (Casos Automatizados)
# =====================================================================
html_facebook = """
<div class="contenedor-fb">
    <div class="anuncio-fb">
        <h1 class="fb-title">REMATO PS5 NUEVA EN CA POR VIAJE</h1>
        <div class="fb-price">S/ 900</div>
    </div>
    <div class="anuncio-fb">
        <h1 class="fb-title">PlayStation 5 Standard + FC24 Juego Incluido</h1>
        <div class="fb-price">S/ 2.650</div>
    </div>
    <div class="anuncio-fb">
        <h1 class="fb-title">PS5 Casi nueva con 1 mando extra Dualsense</h1>
        <div class="fb-price">S/ 1.100</div>
    </div>
</div>
"""

print("==========================================================")
print("  SISTEMA AVANZADO DE DETECCIÓN DE FRAUDE POR CATEGORÍAS  ")
print("==========================================================\n")

# -----------------------------------------------------------------
# PASO 1: Procesar y Segmentar el Mercado Formal (ML)
# -----------------------------------------------------------------
soup_ml = BeautifulSoup(html_mercado_libre, 'html.parser')
titulos_ml = soup_ml.find_all('h2', class_='titulo-prod')
precios_ml = soup_ml.find_all('span', class_='precio-prod')

datos_segmentados = {
    "Consola Estándar": [],
    "Consola + Mando": [],
    "Consola + Juego": []
}

print("[Scraping ML] Extrayendo y segmentando el mercado:")

for i in range(len(titulos_ml)):
    titulo = titulos_ml[i].text
    precio = limpiar_precio(precios_ml[i].text)

    if precio < 1500:
        print(f"   -> 🗑️ [RUIDO DESCARTADO] {titulo[:35]}... (S/ {precio})")
    else:
        categoria = obtener_categoria(titulo)
        # VALIDACIÓN DEFENSIVA: Evita que el programa colapse si devuelve 'No identificado'
        if categoria in datos_segmentados:
            datos_segmentados[categoria].append(precio)
            print(f"   -> 📦 [GUARDADO EN: {categoria}] {titulo[:30]}... | S/ {precio}")
        else:
            print(f"   -> ⚠️ [IGNORADO - NO CATEGORIZADO] {titulo[:30]}... | S/ {precio}")

# Calcular métricas seguras
umbrales_por_categoria = {}
print("\n📊 MÉTRICAS CALCULADAS POR CATEGORÍA:")

for cat, lista_precios in datos_segmentados.items():
    if len(lista_precios) > 0:
        minimo = min(lista_precios)
        promedio = sum(lista_precios) / len(lista_precios)
        critico = minimo * 0.60

        umbrales_por_categoria[cat] = {"minimo": minimo, "promedio": promedio, "critico": critico}
        print(f" * {cat}: Promedio S/{promedio:.0f} | Mínimo Seguro S/{minimo} | Umbral Riesgo S/{critico:.0f}")

# -----------------------------------------------------------------
# PASO 2: Procesar anuncios de Facebook aplicando categorías
# -----------------------------------------------------------------
print("\n" + "="*55)
print(" [Scraping FB] Analizando Fuente 2 (Facebook Marketplace) ")
print("="*55)

soup_fb = BeautifulSoup(html_facebook, 'html.parser')
titulos_fb = soup_fb.find_all('h1', class_='fb-title')
precios_fb = soup_fb.find_all('div', class_='fb-price')

for i in range(len(titulos_fb)):
    titulo_fb = titulos_fb[i].text
    precio_fb = limpiar_precio(precios_fb[i].text)
    categoria_fb = obtener_categoria(titulo_fb)

    print(f"\n-> Anuncio {i+1}: '{titulo_fb}'")
    print(f"   Categoría detectada: {categoria_fb} | Precio publicado: S/ {precio_fb}")

    # VALIDACIÓN DEFENSIVA: Comprobamos si tenemos datos para evaluar esta categoría
    if categoria_fb == "No identificado" or categoria_fb not in umbrales_por_categoria:
        print("   ⚠️ NO SE PUEDE EVALUAR: El anuncio no posee suficientes palabras clave de validación.")
    else:
        umbral_aplicable = umbrales_por_categoria[categoria_fb]["critico"]
        if precio_fb < umbral_aplicable:
            print(f"   🚨 FRAUDE DETECTADO: El precio es demasiado bajo para la categoría '{categoria_fb}'.")
        else:
            print(f"   ✅ ANUNCIO SEGURO: El precio coincide con el mercado para un(a) '{categoria_fb}'.")

# -----------------------------------------------------------------
# PASO 3: INTERFAZ INTERACTIVA INTELIGENTE (A prueba de errores)
# -----------------------------------------------------------------
print("\n" + "="*55)
print(" 🕵️‍♂️ MÓDULO DE CONSULTA MANUAL PARA EL USUARIO 🕵️‍♂️")
print("="*55)

while True:
    print("\n¿Viste otro anuncio sospechoso en redes sociales?")
    entrada_titulo = input("👉 Ingresa el TÍTULO del anuncio (o 'salir'): ")

    if entrada_titulo.lower() == 'salir':
        print("\n¡Gracias por usar el Sistema de Detección de Fraude (LP)!")
        break

    entrada_precio = input("👉 Ingresa el PRECIO que viste (solo números): ")

    try:
        precio_consultado = float(entrada_precio)
        categoria_detectada = obtener_categoria(entrada_titulo)

        # VALIDACIÓN DEFENSIVA GLOBAL
        if categoria_detectada == "No identificado" or categoria_detectada not in umbrales_por_categoria:
            print(f"\n❌ El sistema no pudo reconocer o no tiene datos de referencia para '{entrada_titulo}'.")
            print("   Por favor, asegúrate de incluir palabras clave como 'Consola', 'PS5', 'Mando' o 'Juego'.")
        else:
            datos_mercado = umbrales_por_categoria[categoria_detectada]
            print(f"\n[Analizando...] Producto detectado como: {categoria_detectada}")

            if precio_consultado < datos_mercado["critico"]:
                print(f"🚨 ALERTA CRÍTICA: S/ {precio_consultado} es una estafa casi segura.")
                print(f"   Un(a) {categoria_detectada} original formal no baja de S/ {datos_mercado['minimo']}.")
            else:
                print(f"✅ VERIFICADO: El precio es coherente para un(a) {categoria_detectada}.")

    except ValueError:
        print("❌ Error: Ingresa solo números en el campo de precio.")
```
